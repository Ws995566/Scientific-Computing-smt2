```python
#Soal Nomor 1

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda x, y : x**2 - 4*y
h = 0.01
x = np.arange(0, 1 + h, h)
y0 = 1

y = np.zeros(len(x))
y[0] = y0

for i in range (len(x)-1):
    y[i+1] = y[i] + h*f(x[i], y[i])

exact_y = (1/16) * (1 - np.exp(-4*x)) + (x*3) / 12 -(x*2) / 8 + x / 32

#Ploting
plt.figure(figsize = (12,8))
plt.plot(x, y, label = 'Euler Method')
plt.plot(x, exact_y, label = 'Exact Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
