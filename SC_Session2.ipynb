```python
a = [[8, 3, -3], [-2,-8,5],[3,5,10]]

diag = np.diag(np.abs(a))

off_diag = np.sum(np.abs(a), axis = 1) - diag
if np.all(diag > off_diag):
  print('matrix is diagonaly dominant')
else:
  print('not diagonaly dominant')
```

```python
x1 = 0
x2 = 0
x3 = 0

epsilon = 0.01
converged = False

x_old = np.array([x1,x2,x3])

for k in range (1,50):
  x1 = (14 - 3*x2 + 3*x3) / 8
  x2 = (5 + 2*x1 - 5*x3) / (-8)
  x3 = (-8 - 3*x1 - 5*x2) / (10)
  x_new = np.array([x1,x2,x3])
  dx = np.sqrt (np.dot(x_new - x_old , x_new - x_old))

  print("%d, %.4f, %.4f, %.4f"%(k, x1, x2, x3))

  if dx < epsilon :
    converged = True
    print("converged")
    break

  x_old = x_new

if not converged :
  print("Anda Salah")
```

```python
import numpy as nump
from scipy import optimize
import matplotlib.pyplot as plt

x = np.array([5,4,6,7,5])
y = np.array([10,8,12,14,10])

A = np.vstack([x,np.ones(len(x))]).T

y = y[:,np.newaxis]

alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)

plt.figure(figsize=(10,8))
plt.plot(x,y,'b.', label = 'Data')
plt.plot(x,alpha[0]*x + alpha[1], 'r', label = 'Fit')
plt.xlabel('x')
plt.ylabel('y')
for i, j in zip(x, y):
  plt.text(i, j, '({:.0f},{:.0f})'.format(float(i), float(j)))

plt.show()
```
